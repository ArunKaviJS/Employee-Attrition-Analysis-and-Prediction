```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,classification_report,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import xgboost as xgb
```

```python
models={'Logisticregression':LogisticRegression(),'decisiontreeclassifier':DecisionTreeClassifier(),'randomforestclassifier':RandomForestClassifier(),'xgboost':xgb.XGBClassifier()}
```

```python
def Modeltrain(x_train,x_test,y_train,y_test):
 for name,model in models.items():
     print(f'************************************{name}********************************************')
     model.fit(x_train,y_train)

     print(f'trained with : {name}')

     predict=model.predict(x_test)
     predict_train=model.predict(x_train)

     accuracy_test=accuracy_score(y_test,predict)
     print(f'accuracy_score_test : {accuracy_test}')
     accuracy_train=accuracy_score(y_train,predict_train,)
     print(f'accuracy_score_train:{accuracy_train}')
     print('-----------------------------------------------------------------------------------')
     precision=precision_score(y_test,predict, average='weighted')
     print(f'precision_score_test:{precision}')
     precision_train=precision_score(y_train,predict_train,average='weighted')
     print(f'precision_train:{precision_train}')
     print('-----------------------------------------------------------------------------------')

     recall=recall_score(y_test,predict,average='weighted')
     print(f'recall_test:{recall}')
     recall_train=recall_score(y_train,predict_train,average='weighted')
     print(f'recallscore_train:{recall_train}')
     print('-----------------------------------------------------------------------------------')

     f1=f1_score(y_test,predict,average='weighted')
     print(f'f1_score_test:{f1}')
     f1_train=f1_score(y_train,predict_train,average='weighted')
     print(f'f1score_train:{f1_train}')

     print('-----------------------------------------------------------------------------------')
     #roc=roc_auc_score(y_test,predict,multi_class='ovr')
     #print(f'roc_auc_score_test:{roc}')
     #roc_train=roc_auc_score(y_train,predict_train,multi_class='ovr')
     #print(f'roc_auc_score(:{roc_train}')

     print('-----------------------------------------------------------------------------------')

     cr=classification_report(y_test,predict)
     print(cr)

     cm=confusion_matrix(y_test,predict)
     #print(f'[TN  FP  \n FN TP]:\n{cm}')



```

```python
feature_train,feature_test,target_train,target_test=train_test_split(df,target)
```

```python
Modeltrain(feature_train,feature_test,target_train,target_test)
```

```python
import pandas as pd

# Combine features and target into a single DataFrame
linear = feature_train.copy()
linear["target"] = target_train

# Compute Pearson correlation
correlations = linear.corr()["target"].drop("target").sort_values(key=abs, ascending=False)

# Display top correlated features
print("Top features showing linear correlation with the target:\n")
print(correlations)
```

```python
import numpy as np
import pandas as pd
import joblib

# Load encoders and model
ohe_for_job_sat = joblib.load(r'..\Models\onehotforjobsat.pkl')
label_for_job_sat = joblib.load(r'..\Models\labelforjobsat.pkl')
logistic_performance_rating = joblib.load(r'..\Models\logistic_for_performance_rating.pkl')

# ---------- Collect Inputs ----------
# Numerical inputs
numerical_inputs = {
    "Age": int(input("Enter Age: ")),
    "DistanceFromHome": int(input("Enter Distance from Home: ")),
    "Education": int(input("Enter Education Level: ")),
    "EnvironmentSatisfaction": int(input("Enter Environment Satisfaction: ")),
    "JobInvolvement": int(input("Enter Job Involvement: ")),
    "MonthlyIncome": int(input("Enter Monthly Income: ")),
    "MonthlyRate": int(input("Enter Monthly Rate: ")),
    "PercentSalaryHike": int(input("Enter Percent Salary Hike: ")),
    "RelationshipSatisfaction": int(input("Enter Relationship Satisfaction: ")),
    "WorkLifeBalance": int(input("Enter Work-Life Balance: ")),
    "YearsInCurrentRole": int(input("Enter Years in Current Role: ")),
    
}

# Label encoded inputs
overtime = input("Enter Overtime (Yes/No): ").title()
attrition = input("Enter Attrition (Yes/No): ").title()

numerical_inputs["OverTime"] = label_for_job_sat.transform([[overtime]])[0]


# ---------- Categorical Inputs ----------
cat_input = {
    "BusinessTravel": input("Enter Business Travel (Non-Travel/Travel Frequently/Travel Rarely): ").title(),
    "Department": input("Enter Department: ").title(),
    "EducationField": input("Enter Education Field: ").title(),
    "Gender": input("Enter Gender (Male/Female): ").title(),
    "JobRole": input("Enter Job Role: ").title(),
    "MaritalStatus": input("Enter Marital Status: ").title()
}

# Convert to DataFrame and apply one-hot encoding
cat_df = pd.DataFrame([cat_input])
cat_encoded = ohe_for_job_sat.transform(cat_df)

# ---------- Final Input Vector ----------
num_array = np.array(list(numerical_inputs.values())).reshape(1, -1)
final_input = np.concatenate([num_array, cat_encoded], axis=1)

```

```python
# ---------- Make Prediction ----------
prediction = logistic_performance_rating.predict(final_input)
print("Predicted Output:", prediction[0])
```
